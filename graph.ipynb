```python
import math

# Given volume
V = 100  # cubic kilometers

# Step 1: Calculate the radius of the sphere
r = (3 * V / (4 * math.pi))**(1/3)
S_sphere = 4 * math.pi * r**2
print("Площадь поверхности шара: ", S_sphere)
print("Радиус сферы:", r)

# Step 2: Set a and b to 0.75 * r, calculate c
a = b = 0.75 * r
c = V / ((4 / 3) * math.pi * a**2)
print("Сжатие элипсоида по x и y: ", a,b)
print("Растижение элипсоида пор оси Z:", c)

# Step 3: Calculate the surface area of the ellipsoid
p = 1.6
S_ellipsoid = 4 * math.pi * (
    ((a * b)**p + (a * c)**p + (b * c)**p) / 3
) ** (1 / p)
print("Пощадь поверхности элипсоида:", S_ellipsoid)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to plot the cube
# def plot_cube(ax, size=1):
#     r = [-size / 2, size / 2]
    
#     # Top and bottom faces (z = constant)
#     X, Y = np.meshgrid(r, r)
#     Z_top = np.full_like(X, size / 2)
#     Z_bottom = np.full_like(X, -size / 2)
#     ax.plot_surface(X, Y, Z_top, color='r', alpha=0.3)      # Top face
#     ax.plot_surface(X, Y, Z_bottom, color='r', alpha=0.3)   # Bottom face

#     # Side faces (x = constant)
#     Y_side, Z_side = np.meshgrid(r, r)
#     X_side1 = np.full_like(Y_side, size / 2)
#     X_side2 = np.full_like(Y_side, -size / 2)
#     ax.plot_surface(X_side1, Y_side, Z_side, color='r', alpha=0.3)  # Side face x = size/2
#     ax.plot_surface(X_side2, Y_side, Z_side, color='r', alpha=0.3)  # Side face x = -size/2

    # # Side faces (y = constant)
    # X_side, Z_side = np.meshgrid(r, r)
    # Y_side1 = np.full_like(X_side, size / 2)
    # Y_side2 = np.full_like(X_side, -size / 2)
    # ax.plot_surface(X_side, Y_side1, Z_side, color='r', alpha=0.3)  # Side face y = size/2
    # ax.plot_surface(X_side, Y_side2, Z_side, color='r', alpha=0.3)  # Side face y = -size/2

    # # Highlight the vertices of the cube
    # vertices = np.array([[x, y, z] for x in r for y in r for z in r])
    # ax.scatter(vertices[:, 0], vertices[:, 1], vertices[:, 2], color="b", s=100)

# Function to plot the sphere
# Функция для отображения сферы
def plot_sphere(ax, radius=1, center=(0, 0, 0), color='c', alpha=0.5):
    u = np.linspace(0, 2 * np.pi, 100)
    v = np.linspace(0, np.pi, 100)
    x0, y0, z0 = center
    x = radius * np.outer(np.cos(u), np.sin(v)) + x0
    y = radius * np.outer(np.sin(u), np.sin(v)) + y0
    z = radius * np.outer(np.ones_like(u), np.cos(v)) + z0
    ax.plot_surface(x, y, z, color=color, alpha=alpha)

# Функция для отображения эллипсоида
def plot_ellipsoid(ax, a=1, b=1, c=1, center=(0, 0, 0), color='g', alpha=0.5):
    u = np.linspace(0, 2 * np.pi, 100)
    v = np.linspace(0, np.pi, 100)
    x0, y0, z0 = center
    x = a * np.outer(np.cos(u), np.sin(v)) + x0
    y = b * np.outer(np.sin(u), np.sin(v)) + y0
    z = c * np.outer(np.ones_like(u), np.cos(v)) + z0
    ax.plot_surface(x, y, z, color=color, alpha=alpha)

# Создаем 3D график
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Задаем смещение по оси X, чтобы разместить фигуры рядом
offset = 4
# Create a 3D plot


plot_sphere(ax, radius=r, center=(-offset, 0, 0), color='c', alpha=0.5)

# Отображаем эллипсоид со смещением вправо
plot_ellipsoid(ax, a=a, b=b, c=c, center=(offset, 0, 0), color='g', alpha=0.5)

# Plot the cube with edge length sqrt(2)
#plot_cube(ax, size=np.sqrt(2))

# Set axis labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Set equal aspect ratio
ax.set_box_aspect([1,1,1])

plt.show()
```
