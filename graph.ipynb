```python
import numpy as np
import matplotlib.pyplot as plt

# Function to plot the cube
def plot_cube(ax, size=1):
    r = [-size / 2, size / 2]
    
    # Top and bottom faces (z = constant)
    X, Y = np.meshgrid(r, r)
    Z_top = np.full_like(X, size / 2)
    Z_bottom = np.full_like(X, -size / 2)
    ax.plot_surface(X, Y, Z_top, color='r', alpha=0.3)      # Top face
    ax.plot_surface(X, Y, Z_bottom, color='r', alpha=0.3)   # Bottom face

    # Side faces (x = constant)
    Y_side, Z_side = np.meshgrid(r, r)
    X_side1 = np.full_like(Y_side, size / 2)
    X_side2 = np.full_like(Y_side, -size / 2)
    ax.plot_surface(X_side1, Y_side, Z_side, color='r', alpha=0.3)  # Side face x = size/2
    ax.plot_surface(X_side2, Y_side, Z_side, color='r', alpha=0.3)  # Side face x = -size/2

    # Side faces (y = constant)
    X_side, Z_side = np.meshgrid(r, r)
    Y_side1 = np.full_like(X_side, size / 2)
    Y_side2 = np.full_like(X_side, -size / 2)
    ax.plot_surface(X_side, Y_side1, Z_side, color='r', alpha=0.3)  # Side face y = size/2
    ax.plot_surface(X_side, Y_side2, Z_side, color='r', alpha=0.3)  # Side face y = -size/2

    # Highlight the vertices of the cube
    vertices = np.array([[x, y, z] for x in r for y in r for z in r])
    ax.scatter(vertices[:, 0], vertices[:, 1], vertices[:, 2], color="b", s=100)

# Function to plot the sphere
def plot_sphere(ax, radius=1):
    u = np.linspace(0, 2 * np.pi, 100)
    v = np.linspace(0, np.pi, 100)
    x = radius * np.outer(np.cos(u), np.sin(v))
    y = radius * np.outer(np.sin(u), np.sin(v))
    z = radius * np.outer(np.ones_like(u), np.cos(v))
    ax.plot_surface(x, y, z, color='c', alpha=0.3)

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the sphere
plot_sphere(ax, radius=2)

# Plot the cube with edge length sqrt(2)
plot_cube(ax, size=np.sqrt(2))

# Set axis labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Set equal aspect ratio
ax.set_box_aspect([1,1,1])

plt.show()
```
